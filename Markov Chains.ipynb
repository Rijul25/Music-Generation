```python
def GenerateTable(data,k=4):
    t={}
    
    for i in range(len(data)-k):
        X=data[i:i+k]
        Y=data[i+k]
        print("X %s and Y %s"%(X,Y))
        
        if t.get(X) is None:
            t[X]={}
            t[X][Y]=1
        else:
            if t[X].get(Y) is None:
                t[X][Y]=1
            else:
                t[X][Y] +=1
    return t
```

```python
t = GenerateTable("hello hello helli")
print(t)
```

```python
def convertFreqintoProb(t):
    for kx in t.keys():
        s=float(sum(t[kx].values()))
        for k in t[kx].keys():
            t[kx][k]=t[kx][k]/s
    return t
```

```python
t =convertFreqintoProb(t)
print(t)
```

# Train our Markov Chain

```python
def TrainMarkovChain(text,k=4):
    t = GenerateTable(text,k)
    
    t = convertFreqintoProb(t)
    
    return t
```

```python
model = TrainMarkovChain(text)
```

```python
import numpy as np
fruits=["apple","banana","mango"]
prob = ["0.1","0.8","0.1"]
for i in range(10):
    print(np.random.choice(fruits,p=prob))
```

```python
def sample_text_data(ctx,t,k):
    ctx = ctx[-k:]
    if t.get(ctx) is None:
        return " "
    possible_characters=list(t[ctx].keys())
    possible_values=list(t[ctx].values())
    #print(possible_characters)
    #print(possible_values)
    return np.random.choice(possible_characters,p=possible_values)
```

```python
sample_text_data("comm",model,4)
```

```python
def GenerateText(starting_sent,k=4,maxLen=1000):
    sentence=starting_sent
    ctx=starting_sent[-k:]
    
    for ix in range(maxLen):
        next_prediction=sample_text_data(ctx,model,k)
        sentence += next_prediction
        ctx = sentence[-k:]
    return sentence
```

```python
text = GenerateText("hell",k=4,maxLen=2500)
print(text)
```
